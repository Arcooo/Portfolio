```python
# how to create a tool that does categorical data automatically
# dataset of file
# Takes in the column = for analysis
def statsTotals(dataset, column, colCompr, categ):
    assert(type(dataset) == type(pd.DataFrame()))
    assert(type(column) == type(str()))
    assert(type(categ) == type(list()))

    num_rows = dataset.shape[0] # get number of rows for data
    num_categs = len(categ) # get how many categories were imputed
    
    # create a list to hold the counts
    cntInEachCat = []
    for i in range(num_categs):
        cntInEachCat.append(0)
        cntInEachCat.append(0)

    
    # add an Others category just in case a category not in the Features dic included
#     cntInEachCat.append(0)
####################################################################
#     print(cntInEachCat)
#     print(type(cntInEachCat))
#     f = list(range(2))
#     print(f)
    for row in range(50):
        for cat in range(num_categs):
            if(dataframe_all.loc[row][column] == categ[cat]):   
#                 print(categ[cat]) #when categ = 0 = "Active" else 1 =enrolled
                if(dataframe_all.loc[row][colCompr] > 0):
                    cntInEachCat[cat*num_categs+1] += 1
#                     print(dataframe_all.loc[row][colCompr])
                else:
                    cntInEachCat[cat*num_categs] += 1
    return cntInEachCat

var = statsTotals(dataframe_all, "currentstatus", "revenue_30", ["Active", "Enrolled"])
print(var)
# able to derive this if needed
# Unmanaged Total: 14748
# Onboarding Total: 9986
# Rention Total: 3392
# Unmanaged %: 0.524354689611036
# Onboarding %: 0.3550451539500818
# Rention %: 0.12060015643888217
```
